# K-Means Clustering

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data preparation

```python
# Importing the dataset
df = pd.read_csv('bank.csv')
df = df.drop(['id'], axis=1)
```

```python
df
```

```python
# Function Encoding
def encoding_char(x):
    char_var = list(set(x.columns) - set(x._get_numeric_data().columns))
    for col_names in char_var:
        f = pd.factorize(x[col_names])
        x[col_names] = pd.factorize(x[col_names])[0]
    return(x)
```

```python
# Encoding categorical data
df = encoding_char(df)
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler().fit(df)
df_sc = sc.transform(df)
df_sc = pd.DataFrame(df_sc, columns = ['age','sex','income','married','children','car','mortgage','delays'])
```

## Number of clusters

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(df_sc)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Using the silhouette method to find the optimal number of clusters
from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 6):
  kmeans = KMeans(n_clusters = k).fit(df_sc)  
  preds = kmeans.fit_predict(df_sc)
  sil.append(silhouette_score(df_sc, preds, metric = 'euclidean'))


plt.plot(range(2, 6), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
 print(str(i+2) +":"+ str(sil[i]))
```

## Baseline model

```python
# Fitting K-Means to the dataset
km = KMeans(n_clusters = 4, random_state = 0).fit(df_sc)
km4 = km.predict(df_sc)
pd.DataFrame(km.cluster_centers_, columns = ['age','sex','income','married','children','car','mortgage','delays']).round(2)
```

```python
def plot_2d(X, label_list_cluster):
    plt.figure(figsize=(7, 7))
    plt.subplots_adjust(bottom=0.1)
    c = label_list_cluster
    plt.scatter(X[:, 0], X[:, 1], c=c, cmap='rainbow', label='True Position')
    plt.show()
```

```python
plot_2d(df_sc[['age', 'income']].to_numpy(),km4)
```

## Less features, less clusters

```python
# Less features
X = ['age','income','delays']
```

```python
# Fitting K-Means to the dataset
km_r = KMeans(n_clusters = 2, random_state = 0).fit(df_sc[X])
km2 = km_r.predict(df_sc[X])
km2_centers = pd.DataFrame(km_r.cluster_centers_, columns = ['age','income','delays']).round(2)
km2_centers
```

```python
def plot_2d_cores(X, label_list_cluster, cores):
    plt.figure(figsize=(7, 7))
    plt.subplots_adjust(bottom=0.1)
    c = label_list_cluster
    n = ('orange','green')
    plt.scatter(X[:, 0], X[:, 1], c=c, cmap='rainbow', label='True Position')
    plt.scatter(cores[:, 0], cores[:, 1], c=n, s = 500, marker = '*')
    plt.show()
```

```python
plot_2d_cores(df_sc[['age', 'income']].to_numpy(),km2,km2_centers[['age', 'income']].to_numpy())
```
